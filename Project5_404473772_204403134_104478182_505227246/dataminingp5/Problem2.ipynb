## Definitions

```python
import json
import datetime
from math import ceil
import datetime
import pytz
import numpy as np
import dateutil.parser
import matplotlib.pyplot as plt
pst_tz = pytz.timezone('America/Los_Angeles')

names = ["tweets_#gohawks.txt", "tweets_#gopatriots.txt", \
         "tweets_#nfl.txt", "tweets_#patriots.txt", "tweets_#sb49.txt", "tweets_#superbowl.txt"]
```

## Histogram Function

```python
def tweet_per_hour(name):   
    tweets = []
    mincit = 10e9
    maxcit = -1
    
    for line in open(name, 'r'):
        l = json.loads(line)
        a = l['citation_date']
        if a < mincit:
            mincit = a
        if a > maxcit:
            maxcit = a
        tweets.append(a/3600.0)
    return tweets, mincit/3600.0, maxcit/3600.0
```

```python
tweetsr, mincitr, maxcitr = tweet_per_hour(names[2])
```

```python
bins = np.arange(0, int(maxcitr)-int(mincitr)+1, 1)
tw = [(i - int(mincitr)) for i in tweetsr]
plt.hist(tw, bins)
plt.xlabel('Consequtive Hours', size=15)
plt.ylabel('Number of Tweets', size=15)
plt.title('Hourly Trends of Tweeting: #nfl', size=15)
plt.draw()
plt.savefig('Q2_nfl.png', bbox_inches='tight')
plt.show()
```
